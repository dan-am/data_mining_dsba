```python
import ollama

def generate_synthetic_data(prompt: str, model: str = "Mistral") -> str:
    """
    Generiert synthetische Daten über ollama mittels eines Prompts.
    """
    response = ollama.generate(model=model, prompt=prompt)
    return response

if __name__ == "__main__":
    # Beispiel-Prompt für CSV-Daten
    csv_prompt = """\
Du bist ein KI-System, das synthetische Datensätze generiert.
Erstelle eine CSV mit 50 Zeilen und folgenden Spalten:
- ID (eindeutige Nummer)
- Name (fiktiver Vor- und Nachname)
- Alter (realistische Werte zwischen 18 und 70)
- Geschlecht (männlich, weiblich oder divers)
- Einkommen (realistische Jahresgehälter zwischen 30.000 und 120.000)

Gib ausschließlich die CSV-Daten (inkl. Header) aus.
"""

    csv_data = generate_synthetic_data(prompt=csv_prompt)
    print("CSV-Daten:\n", csv_data)
```

```python
str(csv_data.response)
```

```python
import pandas as pd
from io import StringIO

# Angenommen, csv_data.response ist dein CSV-String:
csv_string = csv_data.response

# 1. In einen DataFrame laden
df = pd.read_csv(StringIO(csv_string))

print(df)
```
